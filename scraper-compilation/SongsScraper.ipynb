```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

browser = webdriver.Chrome("../tools/chromedriver.exe")
URL_LINK = ("https://www.scribd.com/document/427299185/Kapampangan-Song-Lyrics")
browser.get(URL_LINK)
time.sleep(2)

wait = WebDriverWait(browser, 10)

contents = []

while True:
    try:
        SOURCE_HTML = browser.page_source
        soup = BeautifulSoup(SOURCE_HTML, "html.parser")
        content_html_raw = soup.find_all('span', {"class" : "a"})
        
        for content in content_html_raw:
            clean_content = content.get_text()
            if clean_content != ' ':
                print(clean_content)
                contents.append(clean_content)   
    except:
        break
        
    browser.quit()
```

```python
import json
import os.path
import pandas as pd

content_list = []

if os.path.isfile('../json/SongsContent.json') == True:
    with open('../json/SongsContent.json', 'r') as file:
        current_file_content = json.load(file)
        content_list.extend(current_file_content)
else:
    print("File does not exist")

content_list.extend(contents)

df = pd.DataFrame(content_list, columns=["lyrics"])
df
```
