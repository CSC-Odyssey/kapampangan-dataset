# Scrape

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

browser = webdriver.Chrome("../tools/chromedriver.exe")
URL_LINK = ("https://quod.lib.umich.edu/p/philamer/apu0949.0001.001/8?q1=Pampanga+literature&view=text")
browser.get(URL_LINK)
time.sleep(2)

wait = WebDriverWait(browser, 10)

contents = []

while True:
    try:
        wait.until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[2]/div/table[2]/tbody/tr/td/div/span[2]/a/img")))
        content = browser.find_element(By.XPATH, "//*[@id='pvdoccontent']").text
        contents.append({'content': content})

        browser.find_element(By.XPATH, "/html/body/div[2]/div/table[2]/tbody/tr/td/div/span[2]/a/img").click()
        time.sleep(2)
    except:
        break

browser.quit()

# Print the contents
for item in contents:
    print(item['content'])
```

```python
import json
import os.path
import pandas as pd

content_list = []

if os.path.isfile('../json/NovelContent.json') == True:
    with open('../json/NovelContent.json', 'r') as file:
        current_file_content = json.load(file)
        content_list.extend(current_file_content)
else:
    print("File does not exist")

content_list.extend(contents)

df = pd.DataFrame(content_list)
df
```

```python
df.to_csv('ScrapeNovel.csv', index=False)
df
```

```python
import gensim

def sent_to_words(contents):
    for sentence in contents:
        yield(gensim.utils.simple_preprocess(sentence, deacc=True))  # deacc=True removes punctuations


words = list(sent_to_words(df['content'].values.tolist()))
words
```

```python
import json

lst = [element for innerList in words for element in innerList]

with open('../json/NovelContent.json', 'w') as f:
    json.dump(lst, f)
```

```python
words_df = pd.DataFrame(lst, columns=["words"])
words_df.to_csv("../csv/NovelContent.csv")

words_df
```

```python
from collections import Counter

freq_counter = Counter(lst)

freq_table = pd.DataFrame.from_dict(freq_counter, orient='index').reset_index()
freq_table = freq_table.rename(columns={'index':'word', 0:'frequency'})
freq_table.sort_values(by='frequency', ascending=False, inplace=True)

freq_table.to_csv("../csv/NovelContent_Frequency.csv")
freq_table

```
