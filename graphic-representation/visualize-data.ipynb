```python
import pandas as pd

data = pd.read_csv('../csv/freqtable_probability.csv')
data.rename(columns = {'Unnamed: 0' : 'word_id'}, inplace = True)
data[0:10]
```

```python
import matplotlib.pyplot as plt

top_10 = data.sort_values(by = 'frequency', ascending = False)[0:10]
top_10.plot(x = 'word', y = 'frequency', kind = 'bar', rot = 10, fontsize = 10)
```

```python
top_10 = data.sort_values(by = 'probability', ascending = False)[0:10]
top_10.plot(x = 'word', y = 'probability', kind = 'bar', rot = 10, fontsize = 10)
```

```python
df_words = pd.read_csv('../csv/compiled_words.csv')

words = " ".join(str(word) for word in df_words['words'].tolist()).lower()
words[:4]
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

wordcloud = WordCloud(background_color="white", stopwords = STOPWORDS, collocations=True, width=1920, height=1080).generate(words)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
