```python
df_words = pd.read_csv('../csv/compiled_words.csv')

words = " ".join(str(word) for word in df_words['words'].tolist()).lower()
words[:4]
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

wordcloud = WordCloud(background_color="white", stopwords = STOPWORDS, collocations=True).generate(words)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
