```python
import json
import os.path
import pandas as pd

content_list = []

if os.path.isfile('../json/bible_contents.json') == True:
    with open('../json/bible_contents.json', 'r') as file:
        current_file_content = json.load(file)
        content_list.extend(current_file_content)
else:
    print("File does not exist")

content_df = pd.DataFrame(content_list, columns=["sentence"])
content_df
```

```python
content_df.to_csv("../csv/bible_sentences.csv")
content_df
```

```python
import gensim

def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations


words = list(sent_to_words(content_df['sentence'].values.tolist()))
```

```python
import json

lst = [element for innerList in words for element in innerList]

with open('../json/bible_word_list.json', 'w') as f:
    json.dump(lst, f)
```

```python
words_df = pd.DataFrame(lst, columns=["words"])
words_df.to_csv("../csv/bible_words.csv")

words_df
```

```python
from collections import Counter

freq_counter = Counter(lst)

freq_table = pd.DataFrame.from_dict(freq_counter, orient='index').reset_index()
freq_table = freq_table.rename(columns={'index':'word', 0:'frequency'})

freq_table.to_csv("../csv/bible_freqtable.csv")
freq_table

freq_table
```
